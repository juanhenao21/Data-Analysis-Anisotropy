# Critical Values

```python
# Modules

# Data
import numpy as np
from matplotlib import pyplot as plt
from numpy import loadtxt
from scipy.optimize import curve_fit

%matplotlib inline
```

```python
gamma = ['0.037', '0.111', '0.333', '1.000', '3.000', '9.000']
thickness = ['02', '04', '06', '08', '10', '12', '14', '16']
folder = ['11', '13', '17', '19', '23']
```

```python
# Critical Values vs Thickness
for f in folder:
    print('-----------')
    print('Semilla %s' % f)
    print('-----------')

    plt.figure(figsize=(16,9))
    for g in gamma:

        file = '01Critical_Values/Sem%s01aTc_k01_%s.txt' % (f,g)

        Thick, Tc = loadtxt(file, usecols=(0,1,), unpack=True)
        plt.plot(Thick, Tc, '-o', label = "$\\gamma = %s$" % g)
        plt.axis([0, 20, 200, 450])

    plt.legend(loc=4)
    plt.xlabel('Thickness (d)')
    plt.ylabel('Critical Temperature (K)')
    plt.title('Critical Temperature vs Thickness - $k = 0.1 - R_{c} = 3.0 - Sem = %s$' % f)
    plt.grid(True)
    plt.show()
```

```python
# Critical Values vs Thickness (Statistics)
for g in gamma:
    print('-----------')
    print('Gamma %s' % g)
    print('-----------')

    plt.figure(figsize=(16,9))
    Tc_n = np.zeros(shape=(5,8))
    for i, f in enumerate(folder):

        file = '01Critical_Values/Sem%s01aTc_k01_%s.txt' % (f,g)

        Thick, Tc = loadtxt(file, usecols=(0,1,), unpack=True)
        Tc_n[i] = Tc

    Tc_Mean = np.mean(Tc_n, axis=0)
    errors = np.std(Tc_n, axis=0)
    plt.plot(Thick, Tc_Mean, '-o')
    plt.errorbar(Thick, Tc_Mean, yerr=errors)
    plt.axis([0, 20, 200, 450])
    plt.xlabel('Thickness (d)')
    plt.ylabel('Critical Temperature (K)')
    plt.title('Critical Temperature vs Thickness - $k = 0.1 - R_{c} = 3.0 - \\gamma = %s$' % g)
    plt.grid(True)
    plt.show()
```
